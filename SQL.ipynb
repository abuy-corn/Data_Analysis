```python
import sqlite3
from faker import Faker
```

```python

# Connect to the SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('online_store.db')
cursor = conn.cursor()

# Create Customers table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Customers (
    customer_id INTEGER PRIMARY KEY,
    first_name TEXT,
    last_name TEXT,
    email TEXT,
    phone TEXT,
    address TEXT,
    city TEXT,
    state TEXT,
    zip_code TEXT,
    registration_date DATE
)
""")

# Insert sample data
cursor.execute("""
INSERT INTO Customers (customer_id, first_name, last_name, email, phone, address, city, state, zip_code, registration_date)
VALUES
(1, 'John', 'Doe', 'john.doe@example.com', '123-456-7890', '123 Main St', 'Springfield', 'IL', '62701', '2024-01-15'),
(2, 'Jane', 'Smith', 'jane.smith@example.com', '234-567-8901', '456 Elm St', 'Shelbyville', 'IL', '62702', '2024-02-20'),
(3, 'Alice', 'Johnson', 'alice.johnson@example.com', '345-678-9012', '789 Oak St', 'Springfield', 'IL', '62703', '2024-03-10')
""")

# Commit changes and close the connection
conn.commit()
conn.close()
```

```python
# prompt: SELECT LAST NAME ORDER ASC ALSO SELECT EMAIL AND PHONE NUMBER

# Connect to your SQLite database
conn = sqlite3.connect('online_store.db')
cursor = conn.cursor()

# Select last names in ascending order
cursor.execute("SELECT last_name FROM Customers ORDER BY last_name ASC")
last_names = cursor.fetchall()
for row in last_names:
  print(row)

# Select email and phone number
cursor.execute("SELECT email, phone FROM Customers")
email_phone = cursor.fetchall()
for row in email_phone:
  print(row)

# Close the connection
conn.close()
```
